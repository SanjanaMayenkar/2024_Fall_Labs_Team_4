```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import numpy as np
import math
```

## Reading `human_voice.wav` File

```python
samplerate, data = wavfile.read('./sound_files/human_voice.wav')
```

```python
print(samplerate)
```

## Plotting Human Voice Wave

```python
max_time = data.size/samplerate
time_steps = np.linspace(0, max_time, data.size)

plt.plot(time_steps, data)
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
# set the title  
plt.title("Human Voice Wave")
# display the plot
plt.show()
```

```python
data8kHz = data[::6]
```

```python
print(f'Downsampled 8kHz wave (# of samples): {data8kHz.shape[0]}')
```

## Plotting Downsampled Waveform

```python
max_time = data8kHz.size/8000
time_steps = np.linspace(0, max_time, data8kHz.size)

plt.plot(time_steps, data8kHz)
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time (s)")
# set the title  
plt.title("Downsampled Human Voice Wave")
# display the plot
plt.show()
```

# Zooming in on sections

```python
start = 0.5
end = 0.502
```

```python
s, e = time_to_step(start, end, samplerate)
print(s, e)

plt.plot(data[s:e])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Sample")
# set the title  
plt.title("Human Voice Wave (0.5-0.502s)")
# display the plot
plt.show()
```

```python
s, e = time_to_step(start, end, 8000)
print(s, e)
plt.plot(data8kHz[s:e])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Sample")
# set the title  
plt.title("Downsampled Human Voice Wave (0.5-0.502s)")
# display the plot
plt.show()
```

```python
def time_to_step(start, end, sample_rate):
    return math.floor(start * sample_rate), math.floor(end * sample_rate)
```

# Part B

```python
samplerate1, data1 = wavfile.read('./sound_files/M1.wav')
samplerate2, data2 = wavfile.read('./sound_files/M2.wav')
samplerate3, data3 = wavfile.read('./sound_files/M3.wav')
```

Calculating RMS of waves

```python
# square data
data1 = np.array(data1 ** 2, dtype=np.int32)
data2 = np.array(data2 ** 2, dtype=np.int32)
data3 = np.array(data3 ** 2, dtype=np.int32)

# take mean
data1_rms = np.mean(data1 ** 2)
data2_rms = np.mean(data2 ** 2)
data3_rms = np.mean(data3 ** 2)

# take square root
data1_rms = np.sqrt(data1_rms)
data2_rms = np.sqrt(data2_rms)
data3_rms = np.sqrt(data3_rms)

print(data1_rms)
print(data2_rms)
print(data3_rms)
```
